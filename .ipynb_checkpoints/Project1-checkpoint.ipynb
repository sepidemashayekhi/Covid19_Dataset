## Importing the libraries

```python
import numpy as np 
import pandas as pd 
import sklearn 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

## Load and Prepare Data

```python
data=pd.read_csv('country_vaccinations.csv')
data
```

```python
Brazil=data[data.country.str.contains("Brazil")]
Brazil
```

## EDA

```python
def EDA(data):
    print ("data Shape",data.shape)
    print ("data columns ",data.columns)
    print ("Number of misiing values:",data.isnull().sum())
    print (" data info",data.info())
    print ("Dtype data",data.dtypes)
    return data.head()
EDA(Brazil)
```

## Data Preprocessing

```python
Brazil['date']=pd.to_datetime(Brazil['date'],format='%Y-%m-%d')
```

```python
Brazil.drop(['people_fully_vaccinated','daily_vaccinations_raw','people_fully_vaccinated_per_hundred',
         'daily_vaccinations_per_million','people_vaccinated_per_hundred','source_name', 'source_website'],axis=1,inplace=True)
         
```

```python
Brazil.drop(['iso_code','vaccines'],axis=1,inplace=True)
```

```python
Brazil['total_vaccinations'].fillna(Brazil['total_vaccinations'].mean(),inplace=True)
Brazil['people_vaccinated'].fillna(Brazil['people_vaccinated'].mean(),inplace=True)
Brazil['daily_vaccinations'].fillna(Brazil['daily_vaccinations'].mean(),inplace=True)
Brazil['total_vaccinations_per_hundred'].fillna(0,inplace=True)
```

```python
Brazil.isnull().sum()
```

## Strorytelling - Visualization

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=Brazil,x='date',y='total_vaccinations')
```

```python
dff=data.groupby(by='country').sum().sort_values(by='total_vaccinations',ascending=False)[0:20]
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=dff,y=dff.index,x='total_vaccinations')
plt.title("Top 20 Country By total_vaccinations")
plt.show()
```

```python
vacc=data.groupby(by='vaccines').sum().sort_values(by='daily_vaccinations',ascending=False)[0:20]
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=vacc,x=vacc.index,y='daily_vaccinations')
plt.xticks(rotation='vertical')
plt.title("Top 10 Vaccination Distributed by Total Vaccination")

plt.tight_layout(pad=0.4)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(Brazil.corr(),annot=True,cbar=False)
plt.tight_layout(pad=0.5)
plt.show()
```

## Prepare Data for Machine learning

```python
Brazil['day']=Brazil['date'].astype(str).str[8:10].astype(float)
Brazil['month']=Brazil['date'].astype(str).str[5:7].astype(float)
Brazil['year']=Brazil['date'].astype(str).str[0:4].astype(float)
```

```python
x=Brazil.iloc[:,6:9]
y=Brazil.loc[:,'total_vaccinations']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Train your model

```python
from sklearn.linear_model import LinearRegression
linear_Reg=LinearRegression()
linear_Reg.fit(x_train,y_train)
```

```python
from sklearn.model_selection import  cross_val_score
accuracy=cross_val_score(linear_Reg,x_train,y_train,cv=10)
accuracy
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.1)
ridge.fit(x_train,y_train)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
```

```python
from sklearn.model_selection import  GridSearchCV
param=[{'C':[0.25,0.5,0.75,1,1.25],'kernel':['linear']},{'C':[0.25,0.5,0.75,1,1.25],'kernel':['rbf']}]
grid_svr=GridSearchCV(estimator=svr,param_grid=param,scoring='accuracy',cv=10)
```

```python
grid_svr.fit(x_train,y_train)
grid_svr.best_params_
```
